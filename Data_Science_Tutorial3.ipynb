```python
import numpy as np
```

```python
array=np.array([[None,None],[None,None]])
print(array)
print("Q1"+"\n")
```

```python
array2=np.random.randint(10,size=(3,4))
print(array2)
print("Q2"+"\n")
```

```python
array3a=np.ones((2,2))
print(array3a)
```

```python
array3b=np.zeros((2,2))
print(array3b)
print("Q3"+"\n")
```

```python
array4=np.random.randint(10,size=(6,3))
array4_transpose=array4.transpose()
print(array4)
print(array4_transpose)
print("Q4"+"\n")
```

```python
array5a=np.random.randint(10,size=(2,3))
array5b=np.random.randint(10,size=(3,4))
dot5=np.dot(array5a,array5b)
print(dot5)
print("Q5"+"\n")
```
